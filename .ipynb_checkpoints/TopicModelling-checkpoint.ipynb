```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import nltk
from nltk.corpus import stopwords
import textwrap
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
from sklearn.decomposition import NMF
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
nltk.download('stopwords')
```

```python
stops=set(stopwords.words('english'))
```

```python
stops=stops.union({
    'said','would', 'could',' told','also','one','two',
    'mr','new','year',
})
```

```python
df=pd.read_csv('datasets/bbc_text_cls.csv')
```

```python
df.head()
```

```python
vectorizer=CountVectorizer(stop_words=stops)
```

```python
X=vectorizer.fit_transform(df['text'])
```

```python
lda=LatentDirichletAllocation(n_components=10,random_state=12345)
```

```python
lda.fit(X)
```

```python
def plot_top_words(model, feature_names, model_name,n_top_words=10):
    fig, axes = plt.subplots(2, 5, figsize=(30, 15), sharex=True)
    axes = axes.flatten()
    for topic_idx, topic in enumerate(model.components_):
        top_features_ind = topic.argsort()[: -n_top_words - 1 : -1]
        top_features = [feature_names[i] for i in top_features_ind]
        weights = topic[top_features_ind]

        ax = axes[topic_idx]
        ax.barh(top_features, weights, height=0.7)
        ax.set_title(f"Topic {topic_idx +1}", fontdict={"fontsize": 30})
        ax.invert_yaxis()
        ax.tick_params(axis="both", which="major", labelsize=20)
        for i in "top right left".split():
            ax.spines[i].set_visible(False)
        fig.suptitle(model_name, fontsize=40)

    plt.subplots_adjust(top=0.90, bottom=0.05, wspace=0.90, hspace=0.3)
    plt.show()
```

```python
#topics by words
feature_names=vectorizer.get_feature_names_out()
plot_top_words(lda,feature_names,'lda')
```

```python
#documents by topics
Z=lda.transform(X)
```

```python
#pick a random document
#check which topics are associated with it

np.random.seed(0)
i=np.random.choice(len(df))
z=Z[i]

topics=np.arange(10)+1

fig,ax=plt.subplots()
ax.barh(topics,z)
ax.set_yticks(topics)
ax.set_title("True labels: %s" % df.iloc[i]["labels"])
```

```python
def wrap(text):
    return textwrap.fill(text, replace_whitespace=False, fix_sentence_endings=True)

print(wrap(df.iloc[i]["text"]))
```

```python
i=np.random.choice(len(df))
z=Z[i]

fig,ax=plt.subplots()
ax.barh(topics,z)
ax.set_yticks(topics)
ax.set_title("True labels: %s" % df.iloc[i]["labels"])
```

```python
print(wrap(df.iloc[i]["text"]))
```

## using NMF(Non negstive matrix factorization)

```python
vectorizer=TfidfVectorizer(stop_words=stops)
```

```python
X=vectorizer.fit_transform(df['text'])
```

```python
nmf=NMF(n_components=10, beta_loss='kullback-leibler', solver='mu',random_state=0)
```

```python
nmf.fit(X)
```

```python
feature_names=vectorizer.get_feature_names_out()
plot_top_words(nmf,feature_names,'nmf')
```

```python
Z=nmf.transform(X)
```

```python
i=np.random.choice(len(df))
z=Z[i]

fig,ax=plt.subplots()
ax.barh(topics,z)
ax.set_yticks(topics)
ax.set_title("True labels: %s" % df.iloc[i]["labels"])
```

```python
print(wrap(df.iloc[i]['text']))
```
